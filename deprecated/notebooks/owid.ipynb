```python
import pandas as pd
```

```python
df = pd.read_csv('../data_sources/our_world_in_data/owid_ts.csv', 
                 index_col=0, parse_dates=True).rename(
columns={'location': 'country'}).fillna(0, downcast="infer")
df.head()
```

```python
idx = pd.date_range(df.index.min(), df.index.max())
idx.strftime('%Y-%m-%d')
```

```python
for group in df.groupby(['country']):
    #print(group[0])
    x = group[1][['total_cases', 'total_deaths']].reindex(idx, fill_value=0)
    dates = x.index.strftime('%Y-%m-%d')
    list(zip(dates, x['total_cases']))
    (list(zip(dates, x['total_deaths'])))
```
